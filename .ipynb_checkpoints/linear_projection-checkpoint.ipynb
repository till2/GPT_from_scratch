```python
import matplotlib.pyplot as plt
import numpy as np

# plot the unit vector v in magenta
v = (1, 1)
plt.arrow(0, 0, v[0], v[1], head_width=0.03, head_length=0.05, fc='m', ec='m', label="v")

# project v with a linear transformation matrix
A = np.array([
    [-0.7, 1],
    [1, -0.2]])

proj = np.dot(A,v)

# blue arrow
plt.arrow(0, 0, proj[0], proj[1], head_width=0.03, head_length=0.05, fc='b', ec='b', label="Av")

# grey arrow
plt.plot([v[0], proj[0]], [v[1], proj[1]], '--', color='gray')
plt.scatter(v[0], v[1], color='gray')


# Plot the grey arrow head
x0 = proj[0] - (proj[0] - v[0]) * 0.1
y0 = proj[1] - (proj[1] - v[1]) * 0.1

dx = (proj[0] - v[0]) * 0.05
dy = (proj[1] - v[1]) * 0.05

# plt.arrow(v[0], v[1], dx, dy, head_width=0.03, head_length=0.03, fc='gray', ec='gray')
plt.arrow(x0, y0, dx, dy, head_width=0.03, head_length=0.03, fc='gray', ec='gray')

plt.xlim([0, 1.4])
plt.ylim([0, 1.2])
plt.legend()
plt.show()
```
